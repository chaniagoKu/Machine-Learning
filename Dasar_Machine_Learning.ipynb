#**PROBLEM FARMING**

##**Pendahuluan**
Problem framing dalam ML adalah metode kita dalam merumuskan sebuah masalah lantas melihat apakah ada solusi ML yang dapat kita kembangkan.

##**Pola Pikir Machine Learning**
Machine Learning pada dasarnya adalah proses pelatihan sebuah perangkat lunak yang disebut model, yang berguna untuk mempelajari pola pada sebuah dataset. Model ini kemudian dapat digunakan untuk memprediksi data yang belum pernah dikenali sebelumnya.

Kemampuan prediksi ini dapat diimplementasikan ke dalam sebuah produk. Contohnya seperti sebuah sistem yang mampu memprediksi video jenis apa saja yang mungkin disukai pengguna, lalu sistem tersebut akan merekomendasikan video-video serupa lainnya pada  pengguna.

Sebelum kita mulai mendalami berbagai metode machine learning dan teknik implementasinya, kita perlu mengetahui tahapan umum yang dipakai dalam pengembangan ML. Berikut adalah 7 tahapan umum pengembangan ML dan contohnya menurut situs Google Developers.

1. Tentukan tujuan - Misal, saya ingin memprediksi kepadatan arus lalu lintas pada hari tertentu.
2. Buat hipotesis - hipotesis awal saya adalah cuaca mungkin menentukan kepadatan arus lalu lintas.
3. Kumpulkan Data - Kumpulkan data harian arus lalu lintas dan cuaca.
4. Uji hipotesis - Latih model dengan data yang telah dikumpulkan
5. Analisa hasil - Apakah model yang dibuat memiliki prediksi yang baik?
6. Buat kesimpulan - Ternyata cuaca tidak terlalu signifikan menentukan kepadatan arus lalu lintas.
7. Perbaiki hipotesis dan ulangi - Hipotesis yang lebih baik mungkin: hari libur nasional menentukan kepadatan arus lalu lintas

##**Mengidentifikasi Permasalahan yang Cocok untuk Machine Learning**
Tidak setiap masalah dapat dikerjakan oleh Machine Learning. Ada beberapa masalah di mana pemrograman tradisional jauh lebih baik diimplementasi dibanding ML.

###**Sulit Dikerjakan dengan Pemrograman Tradisional**
Sebagai contoh, fitur auto reply pada sebuah aplikasi transportasi daring. Fitur tersebut dapat memprediksi kata-kata yang mungkin ditulis oleh pengguna. Contoh lainnya adalah pada Google Photos, di mana pengguna dapat menemukan foto-foto tertentu tanpa melabeli setiap foto secara manual.

###**Jumlah Data**
Lebih banyak data akan meningkatkan kinerja dari model seperti kemampuan prediksinya. Menurut situs Google Developers, jumlah data yang disarankan untuk model linier sedikitnya ribuan sampel, sedangkan untuk jaringan saraf tiruan minimal dalam skala ratusan ribu sampel. Ketika data lebih sedikit dari yang disarankan, cobalah pendekatan tanpa Machine Learning terlebih dahulu.

###**Automasi Pengambilan Tindakan**
Machine learning bekerja lebih baik dalam membuat action alias tindakan dibanding memberikan kita insight. Ketika kita punya data, dan kita ingin mengetahui pola-pola menarik yang terdapat dalam data tersebut, pendekatan statistik lebih cocok untuk dilakukan.

##**Memutuskan Solusi dalam Machine Learning**
###**Tujuan yang Diharapkan**
Tujuan dari menambahkan model ML ke dalam sebuah sistem atau produk adalah untuk mencapai hasil yang Anda harapkan. Di bawah adalah beberapa contoh model ML yang akan dikembangkan dan tujuan yang ingin dicapai dari model tersebut.



image.png

Kemudian kita perlu tahu apakah metrik yang telah ditetapkan dapat diukur. Contohnya adalah sebuah sistem yang dapat memonitor kesehatan pohon-pohon pada sebuah kebun buah untuk mengurangi jumlah pohon yang mati karena sakit. Metrik yang dapat diukur untuk melihat keberhasilan sistem adalah jumlah dari pohon-pohon sakit yang diprediksi oleh sistem.

Dengan adanya metrik yang jelas, kita dapat mengetahui lebih cepat apakah model kita berhasil atau gagal. Idealnya kita ingin mengetahui secara cepat jika model kita terindikasi gagal. Dengan begitu kita dapat merevisi model kita lebih awal dan mencegah waktu kita terbuang.

Masalah ML dapat dibagi menjadi:
1. Klasifikasi biner
2. Regresi satu dimensi
3. Klasifikasi banyak-kelas, label-tunggal
4. Klasifikasi banyak-kelas, banyak-label
5. Regresi multidimensi

Kemudian setelah merumuskan masalah, Anda coba tentukan apa yang akan diprediksi dari model. Dengan menggabungkan tahapan ini kita dapat meringkas masalah ML yang ingin kita kembangkan seperti “masalah ML yang ingin kita kembangkan adalah masalah klasifikasi 3 kelas, label tunggal, yang bertujuan untuk memprediksi apakah sebuah video masuk dalam 3 kelas tertentu (sangat populer, lumayan populer, dan tidak populer) - satu bulan setelah diunggah”.

##**Identifikasi Sumber Data yang Ada**
Hal selanjutnya yang paling penting ketika ingin mengembangkan ML adalah data yang akan Anda gunakan. Pastikan agar jumlah data yang Anda miliki cukup agar model tidak overfitting, serta label pada dataset tersebut memiliki korelasi terhadap tindakan yang harus dikerjakan sistem ML tersebut.

Terakhir, kita dapat memilih beberapa input/atribut saja pada dataset pada pengembangan model di awal sekali. Pada pengembangan awal kita harus memulai dengan infrastruktur sesederhana mungkin yang tujuannya untuk memvalidasi hipotesis secepat mungkin.

##**Kemampuan Belajar Model**
Tulislah aspek-aspek yang mungkin menjadi penyebab model sulit belajar seperti :
1. Terlalu banyak label negatif dibanding label positif
2. Dataset memiliki sampel yang sangat sedikit
3. Sistem cenderung overfitting dan tidak bisa menggeneralisasi terhadap data-data baru. Contohnya, sistem dapat memprediksi dengan baik data video populer di tahun 2019. Namun dengan data di tahun 2020, sistem tidak dapat memprediksi dengan baik video apa yang populer.
Terakhir, Anda juga harus menilai apakah dataset Anda memiliki bias. Bias ini dapat mempengaruhi  prediksi yang dibuat oleh model dengan tendensi yang buruk. Contohnya kita ingin membuat model pengenalan sepatu.
